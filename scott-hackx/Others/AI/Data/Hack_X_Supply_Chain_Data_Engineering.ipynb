```python
import pandas as pd
```

```python
def combine_csv(fileList, fileName):
  # Create an empty DataFrame to store the merged data
  merged_df = pd.DataFrame()

  # Loop through each CSV file and append its data to the merged DataFrame
  for file in fileList:
      df = pd.read_csv(file)
      merged_df = merged_df.append(df, ignore_index=True)

  # Export the merged DataFrame to a new CSV file
  merged_df.to_csv(f"{fileName}_Data.csv", index=False)

  print("Merged data exported to merged_data.csv")
```

```python
# List of CSV files to merge
csv_files = ["Supplier_Data_1.csv", "Supplier_Data_2.csv", "Supplier_Data_3.csv", "Supplier_Data_4.csv", "Supplier_Data_5.csv"]

combine_csv(csv_files, "Supplier")
```

```python
# List of CSV files to merge
csv_files = ["Retailer_Data_1.csv", "Retailer_Data_2.csv", "Retailer_Data_3.csv", "Retailer_Data_4.csv", "Retailer_Data_5.csv"]

combine_csv(csv_files, "Retailer")
```
